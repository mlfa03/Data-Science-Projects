```python
import pandas as pd
import altair as alt
import datapane as dp
```

```python
dp.login(token='seu-toke')
```

```python
#Getting updated data from Our World in Data
data = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
data.columns
```

```python
data['location'].unique()
```

```python
countries = ['Brazil']
df = data[data.location.isin(countries)]
df
```

```python
#Base for the plots
base = alt.Chart(df).encode(x='date:T', color='location').mark_area(opacity=0.5, stroke='black').interactive().properties(width='container')
```

```python
#y axis for the plots
c1 = base.encode(y='new_cases_smoothed_per_million')
c2 = base.encode(y='new_deaths_smoothed_per_million')
c3 = base.encode(y='total_deaths_per_million')
```

```python
c1.properties(width=500)
```

## Publishing the report

```python
dp.Report(
  dp.DataTable(df, caption=f'Dataset for {countries}'),
  dp.Blocks(dp.Plot(c1, caption='New Cases'), dp.Plot(c2, caption='New Deaths'), columns=2),
  dp.Plot(c3, caption='Total deaths per million')
).publish(name='Covid Panel for Brazil', open=True)
```
